```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lista_nombres = ["Esteban", "Michael", "Carlos", "Sofía", "Fernanda"]
```

```python
lista_edad = list(range(20, 25))
```

```python
np.random.seed(2024)
lista_notas = list(np.random.randint(10, 20, size=5))
```

```python
diccionario = { "Nombres" : lista_nombres,
                "Edad"    : lista_edad,
                "Notas"  : lista_notas
    }
```

```python
df = pd.DataFrame(diccionario)
df
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.bar(x = df.Nombres, height= df.Notas)
ax.set_xlabel('Nombre')
ax.set_ylabel('Notas')
ax.set_title('Notas de los estudiantes de Ciencia de Datos')
plt.show()
```
