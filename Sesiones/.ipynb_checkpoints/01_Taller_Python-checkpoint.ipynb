##  <a id='t1'>1. ¿Qué es la programación?</a>

- La programación implica planificar y estructurar una serie de pasos ordenados para llevar a cabo una tarea específica. Este concepto es aplicable en múltiples situaciones, como organizar una salida, unas vacaciones, o programar los horarios de emisión de programas de televisión y películas en el cine.

- En el ámbito de la informática, la programación se refiere a la creación de programas o aplicaciones mediante la escritura de un código fuente. Este código consiste en un conjunto de instrucciones que el ordenador debe seguir para ejecutar un programa. 

- El propósito principal de la programación es establecer instrucciones para que un ordenador pueda ejecutar sistemas, programas y aplicaciones que sean eficientes, accesibles y fáciles de usar.

- Los programas de software generalmente siguen algoritmos, que son conjuntos de instrucciones organizadas y relacionadas entre sí, permitiendo al software operar en equipos de computación.

Fuente: [Concepto.de](https://concepto.de/programacion/#ixzz8eC6Zj08G)

##  <a id='t1.1.'>1.1 Lenguajes de programación</a>

- Un lenguaje de programación es un lenguaje artificial preestablecido compuesto por signos, palabras y símbolos, que permite la comunicación entre el programador y el ordenador.

- Las instrucciones para ejecutar aplicaciones y programas están escritas en un lenguaje de programación y luego se traducen a un lenguaje de máquina que el hardware del equipo puede interpretar y ejecutar.

- El código fuente consiste en líneas de texto que, en lenguaje de programación, indican las acciones que debe realizar el ordenador. Este código se crea, diseña, codifica, mantiene y depura a través del proceso de programación.

- Existen diversos lenguajes de programación, como Java, Perl y Python, que utilizan distintos programas para volcar las instrucciones. Estos lenguajes evolucionan y se expanden con el tiempo.

##  <a id='t1.2.'>1.2 Tipos de lenguajes de programación</a>
Los lenguajes de programación se pueden clasificar en diferentes niveles según su cercanía al lenguaje de máquina y su facilidad de uso para los programadores

#### Lenguaje de Bajo Nivel o Lenguaje de Máquina
- Es el lenguaje más básico y consiste en instrucciones directamente ejecutables por el hardware del ordenador. Está compuesto por cadenas de bits (0s y 1s).
- Características: Muy eficiente en términos de ejecución, pero extremadamente difícil de programar y depurar.
- Ejemplo: 
Código binario (1 y 0)

#### Lenguaje ensamblador
- Es un paso por encima del lenguaje de máquina y utiliza abreviaturas o mnemonics en lugar de números binarios.
- Características: Combina la eficiencia y control de un lenguaje de bajo nivel con características de lenguajes de alto nivel.
- Ejemplo: 
```MOV AX, 1``` (mueve el valor 1 al registro AX).

#### Lenguaje de alto nivel
- Estos lenguajes están diseñados para ser fáciles de leer y escribir por los humanos. Permiten expresar los algoritmos de una manera adecuada a la capacidad cognitiva humana, en lugar de la capacidad con que las máquinas lo ejecutan.
- Características: Permiten una máxima flexibilidad al programador a la hora de abstraerse o de ser literal.
- Ejemplo:

Python: ```print("Hello, World!")```

Java: ```System.out.println("Hello, World!");```

C#: ```Console.WriteLine("Hello, World!");```

Para el aprendizaje automático, Scikit-learn ofrece una suite completa de algoritmos, y TensorFlow y PyTorch son frameworks líderes para el aprendizaje profundo.

3. **Comunidad y Soporte**: Python tiene una comunidad grande y activa, lo que significa que hay una gran cantidad de recursos, tutoriales y documentación disponibles. Este soporte comunitario ayuda a los nuevos usuarios a aprender y resolver problemas, mientras que los usuarios avanzados se benefician del desarrollo y las mejoras continuas.

4. **Integración y Automatización**: La capacidad de Python para integrarse con otras tecnologías y herramientas, como SQL para la gestión de bases de datos o Apache Spark para el procesamiento de grandes volúmenes de datos, lo convierte en una herramienta versátil para los científicos de datos. Además, Python es excelente para la automatización de tareas repetitivas y flujos de trabajo en proyectos de datos.

5. **Adopción en la Industria**: Python es ampliamente adoptado tanto en el ámbito académico como en la industria, lo que asegura que las habilidades aprendidas en Python sean altamente transferibles y valoradas en el mercado laboral. Es comúnmente utilizado por grandes empresas y organizaciones, lo que subraya su relevancia e importancia en el campo.

Dadas estas ventajas, estudiar Python puede proporcionar una base sólida para una carrera en ciencia de datos, permitiéndote manejar una variedad de tareas de manera eficiente y efectiva.

| `!=`     | No igual a         | `3 != 2`   |  `True`   |

### 5.3. <a id='5.3.'>Diccionarios</a>

Un diccionario en Python es **una estructura de datos que permite almacenar pares de clave-valor**. Cada elemento del diccionario tiene una clave única que se utiliza para acceder a su valor correspondiente. Es como un libro de palabras y definiciones, donde las claves son las palabras y los valores son las definiciones. Los diccionarios se crean utilizando **llaves {}** y son muy eficientes para buscar valores asociados a una clave específica.

````python
persona_1 = {'nombre':'José', 'apellido':'Gonzáles', 'edad':35, 'email':'jgonzales@gmail.com' }
````

Podemos acceder a la información que necesitamos a través de las llaves

````python
persona_1['nombre']
````

Puede haber más de un valor por cada llave

````python
personas = { 'nombres' :['José', 'Antonia', 'María'],
             'apellido':['Gonzáles', 'Méndez', 'Rojas'],
             'edad'    :[35, 35, 35],
             'email'   :['jgonzales@gmail.com', 'amendez@gmail.com', 'mrojas@gmail.com'] }
````

````python
personas['edad']
````

Podemos añadir un nuevo par clave-valor

````python
persona_1.update( { 'Residencia' :  'Lima' } )
persona_1
````

Es posible emparejar listas en pares clave-valor

````python
# Definir listas
ciudades      = ['Fray Martin','Santa Rosa de Puquio','Cuchicorral','Santiago de Punchauca' ]
codigo_postal = [15001,15003,15004,15006]

# Emparejar las listas
diccionario_1 = dict( zip( ciudades , codigo_postal ) )

# Visualizar el resultado
diccionario_1
````

##  <a id='t6'>6. Control de flujo</a>

Hasta ahora hemos elaborado scripts que se ejecutan de arriba hacia abajo y de izquierda a derecha; cuando se ejecuta el la última línea, el programa termina. Sin embargo, existen herramientas en Python que permiten alterar el flujo normal de un programa. El uso de estas herramientas es denominado control de flujo.

### 6.1. <a id='6.1.'>Sentencias condicionales</a>

Este tipo de sentencias nos permiten alterar el flujo de un programa mediante el cumplimiento de una condición.

| Sentencia condicional    | Operacion realizada                                              |
|--------------------------|------------------------------------------------------------------|
| `if`                     | Se usa para ejecutar un bloque de código si una condición es verdadera                            |
| `else`                   | Se usa en conjunto con if para ejecutar un bloque de código cuando la condición en el if es falsa |
| `elif`                   | Se usa para verificar múltiples condiciones en orden y ejecutar el bloque de código de la primera condición verdadera                     |

````python
# Ejemplo 1
temperatura = 35

if temperatura > 30:
    print("Hace calor, lleva protector solar.")
````

````python
# Ejemplo 2
temperatura = 25

if temperatura > 30:
    print("Hace calor, lleva protector solar.")
else:
    print("No hace calor, descuida.")
````

````python
# Ejemplo 3
temperatura = 25

if temperatura > 30:
    print("Hace calor, lleva protector solar.")
elif temperatura > 20:
    print("No hace calor, descuida")
else:
    print("Hace frío, lleva un abrigo")
````

### 6.2. <a id='6.2.'>Sentencias de bucle</a>

Se usan para repetir una acción una o más veces

| Sentencia de bucle       | Operacion realizada                                              |
|--------------------------|------------------------------------------------------------------|
| `for`                     | Se utiliza para iterar sobre una secuencia (como una lista, tupla o cadena) o un rango de números y ejecutar un bloque de código para cada elemento en esa secuencia.                            |
| `while`                   |  Se utiliza para repetir un bloque de código mientras una condición sea verdadera. El código se ejecuta una y otra vez hasta que la condición sea falsa. |

````python
# Ejemplo 1:
lista_estaturas = [165, 172, 171, 182, 168, 165]

for estatura in lista_estaturas:
    estatura_metros = estatura/100
    print("Su estatura es " + str(estatura_metros) + " metros")
````

````python
lista_numeros = [1, 2, 3, 4, 5, 7, 8, 9, 10]

for n in lista_numeros:
  suma = n + 10
  print("el numero es " + str(suma))
````

````python
vida_jugador = 10

while vida_jugador > 0:
    print("¡Te quedan", vida_jugador, "puntos de vida!")
    ataque_enemigo = int(input("Enemigo ataca. ¿Cuánto daño recibes? "))
    vida_jugador = vida_jugador - ataque_enemigo

print("¡Game over! Tu personaje ha sido derrotado.")
````

````python
vida_jugador = 10.0  #cambiar a float

while vida_jugador > 0:
    print("¡Te quedan", vida_jugador, "puntos de vida!")

    # cambiar input a float
    ataque_enemigo = float(input("Enemigo ataca. ¿Cuánto daño recibes? "))

    vida_jugador = vida_jugador - ataque_enemigo

print("¡Game over! Tu personaje ha sido derrotado.")
````

##  <a id='t7'>7. Funciones</a>

Las funciones en Python son **bloques de código**  que realizan tareas específicas y pueden ser reutilizados en diferentes partes de un programa. Funcionan como "máquinas" que toman datos de entrada (llamados argumentos o parámetros), los procesan y devuelven un resultado. Aquí hay una explicación breve:

````python
def saludar(nombre):
    mensaje = "Hola, " + nombre
    return mensaje
````

````python
saludar("Lucero")
````

La estructura básica de una función es la siguiente:

````python
def calcular_imc(peso, estatura):
    '''
    Objetivo:
        - Calcular el IMC de una persona dado su peso y altura

    Input:
        - peso:     peso en kilogramos
        - estatura: estatura en centrímetros

    Output:
        - IMC calculado
    '''
    estatura_cm = estatura/100
    imc         = peso/(estatura_cm)**2
    imc         = round(imc, 3)
    return imc
````

````python
calcular_imc(65,160)
````

Un ejemplo adicional de función:

````python
def pobreza_hogares( n_personas, n_habitaciones, ingreso ):

    '''
    Objetivo:
        - Determinar si un hogar es pobre

    Input:
        - n_personas:     número de personas que componen regularmente el hogar
        - n_habitaciones: número de habitaciones del hogar
        - ingreso:        ingreso monetario bruto mensual a nivel de hogar

    Procedimiento:
        - Se calcula el ratio de personas por habitación. Si este ratio es menor a 2.4,
          se considera que existe hacinamiento en el hogar.
        - Se calcula el ingreso per cápita mensual. Si el valor de este indicador es
          mayor a 450, se considera que el ingreso es suficiente (no pobre).
        - Si se cumplen las dos condiciones: no hacinamiento e ingreso suficiente,
          el hogar se considera "no pobre". Todos los demás casos posibles sí
          implican que el hogar es pobre.

    Output:
        - Un veredicto que determina si el hogar es pobre.
    '''

    ratio_habs         = n_personas / n_habitaciones
    ingreso_pc         = ingreso    / n_personas
    no_hacinamiento    = ratio_habs < 2.4
    ingreso_suficiente = ingreso_pc > 450

    no_pobre = no_hacinamiento and ingreso_suficiente

    if no_pobre  == True:
        veredicto = "No es un hogar pobre"
    else:
        veredicto = "Sí es un hogar pobre"

    return veredicto
````

````python
# Hogar 1

n_personas_1     = 5
n_habitaciones_1 = 3
ingreso_1        = 1800

pobreza_hogares(n_personas_1, n_habitaciones_1, ingreso_1)
````

````python
# Hogar 2

n_personas_2     = 3
n_habitaciones_2 = 3
ingreso_2        = 2500

pobreza_hogares(n_personas_2, n_habitaciones_2, ingreso_2)
````

````python
# Hogar 3

n_personas_3     = 10
n_habitaciones_3 = 20
ingreso_3        = 4000

pobreza_hogares(n_personas_3, n_habitaciones_3, ingreso_3)
````
