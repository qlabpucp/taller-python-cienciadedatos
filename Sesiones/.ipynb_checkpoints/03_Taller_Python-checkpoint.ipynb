## <a id='10.1.'>10.1. Visualizacion con Matplotlib</a>
Matplotlib es una de las bibliotecas más utilizadas y versátiles para la visualización de datos en Python. Es conocida por su capacidad para crear gráficos estáticos, animados e interactivos con un alto grado de personalización. Matplotlib permite a los usuarios generar una amplia variedad de gráficos, incluyendo gráficos de líneas, barras, histogramas, dispersión y gráficos tridimensionales, entre otros.

**Características Clave de Matplotlib**
- Versatilidad: Puede generar una amplia gama de tipos de gráficos.
- Personalización: Ofrece un control detallado sobre cada aspecto de la visualización, desde los colores y estilos de las líneas hasta las etiquetas y leyendas.
- Compatibilidad: Se integra bien con otras bibliotecas de análisis de datos en Python, como NumPy y Pandas.

| Método           | Descripción                                                                 |
|------------------|-----------------------------------------------------------------------------|
| `plt.plot(x, y)` | Crea un gráfico de líneas con `x` e `y` como datos.                         |
| `plt.scatter(x, y)` | Crea un gráfico de dispersión con `x` e `y` como datos.                  |
| `plt.bar(x, height)` | Crea un gráfico de barras con `x` como etiquetas de barras y `height` como altura de las barras. |
| `plt.hist(x, bins)` | Crea un histograma de los datos `x` dividido en `bins` contenedores.     |
| `plt.pie(x, labels)` | Crea un gráfico de pastel con `x` como valores y `labels` como etiquetas de los segmentos. |
| `plt.boxplot(x)` | Crea un diagrama de caja para mostrar la distribución de los datos `x`.     |

### Interfaz de Pyplot
```plt.subplots()```: Esta función crea una figura (fig) y uno o más subplots (ejes ax). Es una forma conveniente de crear una figura y un conjunto de subplots en una sola línea de código.

```fig```: Representa la figura creada. Es el contenedor de nivel más alto que contiene todos los elementos del gráfico. Puedes pensar en fig como un lienzo en blanco sobre el cual se dibujan uno o más gráficos.

```ax```: Representa uno o más ejes dentro de la figura. Un eje es una región de la figura en la que se dibuja el gráfico.

````python
fig, ax = plt.subplots()
plt.show()
````

## <a id='10.2.'>10.2. Customizar tus gráficos</a>

### Añadir un marcador

````python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(seattle_station["DATE"], seattle_station["MLY-TAVG-NORMAL"],
       marker="o")
ax.set_title("Temperatura promedio en Seattle")
ax.set_xlabel("Mes")
ax.set_ylabel("Farenheit (F°)")
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(seattle_station["DATE"], seattle_station["MLY-TAVG-NORMAL"],
       linestyle="--")
ax.set_title("Temperatura promedio en Seattle")
ax.set_xlabel("Mes")
ax.set_ylabel("Farenheit (F°)")
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(seattle_station["DATE"], seattle_station["MLY-TAVG-NORMAL"],
       marker="*",linestyle="--", color="darkblue")
ax.set_title("Temperatura promedio en Seattle")
ax.set_xlabel("Mes")
ax.set_ylabel("Farenheit (F°)")
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(seattle_station["DATE"], seattle_station["MLY-TAVG-NORMAL"],
       marker="x",linestyle="--", color="black")
ax.set_title("Temperatura promedio en Seattle")
ax.set_xlabel("Mes")
ax.set_ylabel("Farenheit (F°)")
plt.show()
````

## <a id='10.3.'>10.3. Plots de área</a>

````python
data = pd.read_excel('../datasets/Canada.xlsx', sheet_name='Canada by Citizenship' )
data
````

````python
data.columns
````

````python
data.columns = list(map(str, data.columns))
data.columns
````

````python
data.set_index('OdName', inplace=True)
data.head()
````

````python
data['Total'] = data.sum(axis=1)
data.head(3)
````

````python
years = list(map(str, range(1980, 2014)))
print(years)
````

````python
data.sort_values(by='Total', ascending=False, axis=0, inplace=True)
data_top5 = data[years].head(5)
data_top5
````

````python
data_top5 = data_top5.transpose()
data_top5.head()
````

````python
data_top5.plot( figsize= [12,8] )
plt.title('Evolucion de Inmigracion a Canada de los 5 paises mas aportan inmigrantes',fontdict={'fontsize':16})
plt.xlabel('Año')
plt.ylabel('Numero de Inmigrantes')
plt.grid(True)
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(15, 8))

# Realizamos el gráfico de área apilada en los ejes
data_top5.plot(kind='area', stacked=True, alpha=0.2, ax=ax)

# Añadimos título y etiquetas
ax.set_title('Evolución de Inmigración a Canadá de los 5 países que más aportan inmigrantes')
ax.set_ylabel('Número de Inmigrantes')
ax.set_xlabel('Año')

# Añadimos una cuadrícula
ax.grid(True)

plt.show()
````

````python
np.random.seed(0)

data_area = {
    'Año': range(2010, 2021),
    'Comida': np.random.randint(500, 800, size=11),
    'Hospedaje': np.random.randint(600, 800, size=11),
    'Transporte': np.random.randint(250, 450, size=11),
    'Salud': np.random.randint(180, 300, size=11),
    'Educación': np.random.randint(140, 260, size=11)
}

df_gastos = pd.DataFrame(data_area)
df_gastos.set_index('Año', inplace=True)
df_gastos.head(3)
````

## <a id='10.4.'>10.4. Plots de barras</a>

* <b>kind=bar</b> a la funcion plot  genera un plot de barras verticales.  
* <b>kind=barh</b> a la funcion plot  genera un plot de barras horizontales.

### ```plt.bar(x, y)```

````python
data.head(5)
````

````python
data_chile = data[data.index=="Chile"]
data_chile = data_chile.drop("Total", axis=1)
data_chile = data_chile.T
data_chile.head()
````

````python
fig, ax = plt.subplots(figsize=(12,6))

ax.bar(x = data_chile.index, height= data_chile["Chile"], color="darkred")
ax.set_xlabel('Año')
ax.set_ylabel('Numero de Inmigrantes')
ax.set_title('Inmigrantes de Chile hacia Canada entre 1980 a 2013')
ax.tick_params(axis="x", rotation=90)
plt.show()
````

````python
data_peru = data[data.index=="Peru"]
data_peru = data_peru.drop("Total", axis=1)
data_peru = data_peru.T
data_peru.head()
````

````python
fig, ax = plt.subplots(figsize=(12,6))

ax.bar(x = data_peru.index, height= data_peru["Peru"], color="darkblue")
ax.set_xlabel('Año')
ax.set_ylabel('Numero de Inmigrantes')
ax.set_title('Inmigrantes de Peru hacia Canada entre 1980 a 2013')
ax.tick_params(axis="x", rotation=90)
plt.show()
````

````python
data_peru_chile = data[(data.index=="Peru") | (data.index=="Chile")]
data_peru_chile = data_peru_chile.drop("Total", axis=1)
data_peru_chile = data_peru_chile.T
data_peru_chile.head()
````
